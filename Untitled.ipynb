```python
## function to replace string values with integer values

def convert(listvalues):
    values = {}
    for i in range(0,len(listvalues)):
        values[listvalues[i]]=i
    return values    
values = convert(list(data['ocean_proximity'].unique()))

numData = data.copy(deep=False)

print numData

numData['ocean_proximity'] = [values[prox] for prox in data['ocean_proximity']]

print numData
```

Inference :

## Performing Linear Regression

```python
target = pd.DataFrame(numData['median_house_value'])
target.columns = ['median_house_value']
features = numData.drop(['median_house_value'],axis=1)
```

```python
print target
```

```python
print features
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(features, target, test_size=0.2)
```

```python
print xtrain.shape
print xtest.shape
print ytrain.shape
print ytest.shape
```

```python
## Standardizing the data

stanxtrain = preprocessing.scale(xtrain)
stanxtest = preprocessing.scale(xtest)
```

```python
## Performing linear regression

regr = linear_model.LinearRegression()
regr.fit(stanxtrain,ytrain)
targetPred = regr.predict(stanxtest)
```

```python
## Calculating error

error = mean_squared_error(ytest,targetPred)
print error

## Root Mean Squared Error

error = sqrt(error)
print error
```

```python
## Plotting the difference

plot.scatter(targetPred,ytest,color="blue")
plot.show()
```
